```python
import numpy as np
import pandas as pd
import importlib
import ann_utils
importlib.reload(ann_utils)
import matplotlib.pyplot as plt
```

```python
def bar_chart(dat,n=4,loc1=60,loc2=8,fig1=14,fig2=7):
    img_count=ann_utils.count_img(dat)
    label_count=ann_utils.count_label(dat)
    
    fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(fig1,fig2))
    ax[0].bar(label_count['Label'],label_count['No of annotations'],color="green")
    ax[0].set_title("Number of annotations per Label",size=14)
    for i in range(n):
        ax[0].text(i,label_count['No of annotations'][i]+loc1,label_count['No of annotations'][i],ha='center')

    ax[1].bar(img_count['Label'],img_count['No of Images'],color="green")
    ax[1].set_title("Number of images per Label",size=14)

    for i in range(n):
        ax[1].text(i,img_count['No of Images'][i]+loc2,img_count['No of Images'][i],ha='center')
```
